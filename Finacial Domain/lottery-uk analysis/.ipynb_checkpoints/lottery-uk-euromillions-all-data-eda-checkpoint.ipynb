```python
import base64
import io
import os

from IPython.display import display, HTML
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
def get_csv_files_from_directory(directory, extensions=["csv"]):
    """Get a list of CSV files from the specified directory."""
    datasets = []
    for dirname, _, filenames in os.walk(directory):
        for filename in filenames:
            if filename.split(".")[-1] in extensions:
                datasets.append(os.path.join(dirname, filename))
    return datasets


def load_csv_file(file_path, index_col=None):
    """Load a CSV file into a DataFrame."""
    try:
        df = pd.read_csv(file_path, index_col=index_col)
        return df
    except Exception as e:
        print(f"Error loading {file_path}: {e}")
        return None


directory = "/kaggle/input"
csv_files = get_csv_files_from_directory(directory)

if not csv_files:
    print("No CSV files found.")

for file in csv_files:
    if "eum_df.csv" in file:  # Ensure only the desired file is processed
        df = load_csv_file(file, index_col="DrawNumber")
        if df is not None:
            print(f"Loading df from {file}.")
            try:
                df["DrawDate"] = pd.to_datetime(df["DrawDate"], format="%d-%m-%Y")

            except Exception as e:
                print(f"Error processing {file}: {e}")
    else:
        print(f"Skipping {file} as it is not 'eum_df.csv'")

df.describe(include="all")

print(
    f"There is {len(df)} draws",
    f"with first from {df['DrawDate'].iloc[-1].date()} ",
    f"and last from {df['DrawDate'].iloc[0].date()}",
)
```

```python
# Plot the distribution of each ball

# Set the style for the plots
sns.set(style="darkgrid")

# Define colors for each ball chart
ball_colors = ["blue", "green", "red", "purple", "orange"]
star_colors = ["gold", "brown"]
colors_11 = ["lightgoldenrodyellow", "wheat"]
colors_12 = ["darkgoldenrod", "saddlebrown"]

labels = [
    "1st number drawn",
    "2nd number drawn",
    "3rd number drawn",
    "4th number drawn",
    "5th number drawn",
]

# Plot the distribution of each ball
fig, axes = plt.subplots(1, 5, figsize=(25, 5))

for i, ball in enumerate(["Ball1", "Ball2", "Ball3", "Ball4", "Ball5"]):
    ax = axes[i]
    sns.histplot(df[ball], bins=range(1, 52), kde=False, ax=ax, color=ball_colors[i])

    ax.set_title(f"Distribution of {ball}")
    ax.set_xlabel(labels[i])
    ax.set_ylabel("Frequency")
    ax.set_xticks(range(1, 51))
    ax.set_xlim(1, 50)
    ax.tick_params(axis="x", rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Filter the DataFrame based on the date
df_pre_2016 = df[df["DrawDate"] < pd.Timestamp("2016-09-24")]
df_post_2016 = df[df["DrawDate"] >= pd.Timestamp("2016-09-24")]

# Plot the distribution of each Lucky Star for different periods
fig, axes = plt.subplots(3, 2, figsize=(15, 15))
labels = ["1st LuckyStar drawn", "2nd LuckyStar drawn"]

# Overall distribution
for i, star in enumerate(["LuckyStar1", "LuckyStar2"]):
    ax = axes[0, i]
    sns.histplot(df[star], bins=range(1, 14), kde=False, ax=ax, color=star_colors[i])
    ax.set_title(f"Distribution of {star}")
    ax.set_xlabel(labels[i])
    ax.set_ylabel("Frequency")
    ax.set_xticks([x + 0.5 for x in range(1, 13)])
    ax.set_xticklabels(range(1, 13))
    ax.set_xlim(1, 13)
    ax.tick_params(axis="x", rotation=90)

# Distribution before 24 September 2016
for i, star in enumerate(["LuckyStar1", "LuckyStar2"]):
    ax = axes[1, i]
    sns.histplot(
        df_pre_2016[star], bins=range(1, 13), kde=False, ax=ax, color=colors_11[i]
    )
    ax.set_title(f"Distribution of {star} (Before 24 Sep 2016)")
    ax.set_xlabel(labels[i])
    ax.set_ylabel("Frequency")
    ax.set_xticks([x + 0.5 for x in range(1, 12)])
    ax.set_xticklabels(range(1, 12))
    ax.set_xlim(1, 12)
    ax.tick_params(axis="x", rotation=90)

# Distribution after 24 September 2016
for i, star in enumerate(["LuckyStar1", "LuckyStar2"]):
    ax = axes[2, i]
    sns.histplot(
        df_post_2016[star], bins=range(1, 14), kde=False, ax=ax, color=colors_12[i]
    )
    ax.set_title(f"Distribution of {star} (After 24 Sep 2016)")
    ax.set_xlabel(labels[i])
    ax.set_ylabel("Frequency")
    ax.set_xticks([x + 0.5 for x in range(1, 13)])
    ax.set_xticklabels(range(1, 13))
    ax.set_xlim(1, 13)
    ax.tick_params(axis="x", rotation=90)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, HTML


def calculate_column_frequencies(df, column):
    total_counts = df.shape[0]
    counts = df[column].value_counts()
    likelihoods = {number: count / total_counts for number, count in counts.items()}
    return pd.DataFrame.from_dict(likelihoods, orient='index', columns=['Likelihood']).sort_index()

def plot_likelihoods_with_table(likelihoods_df, title):
    fig, ax = plt.subplots(figsize=(6, 4))
    likelihoods_df.plot(kind='bar', legend=False, ax=ax)
    ax.set_title(title)
    ax.set_xlabel('Number')
    ax.set_ylabel('Likelihood')
    plt.close(fig)  
    
    fig_html = f'<img src="data:image/png;base64,{plt_to_base64(fig)}"/>'
    
    sorted_likelihoods_df = likelihoods_df.sort_values(by='Likelihood')

    html_table = sorted_likelihoods_df.to_html()
    display(HTML(f"""
    <div style="display: flex; align-items: flex-start;">
        <div style="width: 50%; padding-right: 20px;">{fig_html}</div>
        <div style="width: 50%; max-height: 400px; overflow-y: auto; border: 1px solid black;">{html_table}</div>
    </div>
    """))


def plt_to_base64(fig):
    buf = io.BytesIO()
    fig.savefig(buf, format='png')
    buf.seek(0)
    return base64.b64encode(buf.read()).decode('utf-8')
```

```python
ball_columns = ['Ball1', 'Ball2', 'Ball3', 'Ball4', 'Ball5']
lucky_star_columns = ['LuckyStar1', 'LuckyStar2']

ball_likelihoods_dfs = {col: calculate_column_frequencies(df, col) for col in ball_columns}
lucky_star_likelihoods_dfs = {col: calculate_column_frequencies(df, col) for col in lucky_star_columns}
```

```python
for col, likelihoods_df in ball_likelihoods_dfs.items():
    plot_likelihoods_with_table(likelihoods_df, f'Likelihood of Numbers in {col}')

for col, likelihoods_df in lucky_star_likelihoods_dfs.items():
    plot_likelihoods_with_table(likelihoods_df, f'Likelihood of Numbers in {col}')
```
