# Importing the Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the Dataset

```python
file_path = 'USA.csv'
usa = pd.read_csv(file_path)
```

# Convert 'Last' column to numeric

```python
usa['Last'] = pd.to_numeric(usa['Last'], errors='coerce')
```


# Select the top 50 companies based on 'Last' column




```python
top_50_companies = usa.nlargest(50, 'Last')
```

# Visualization 1: Horizontal bar plot for Last recorded stock prices for the top 50 companies

```python
plt.figure(figsize=(12, 14))
sns.barplot(x='Last', y='Name', data=top_50_companies, palette='viridis')
plt.title('Last Recorded Stock Prices for Top 50 S&P 500 Companies')
plt.xlabel('Last Recorded Stock Price')
plt.ylabel('Company')
plt.show()
```


# Visualization 2: Box plot for trading volumes of the top 50 companies


```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Vol', data=top_50_companies, color='skyblue')
plt.title('Box Plot of Trading Volumes for Top 50 Companies')
plt.xlabel('Trading Volume')
plt.show()
```


# Visualization 3: Scatter plot between High and Low stock prices for the top 20 companies


```python
#Choosing Top 5 Comapanies
top_20= top_50_companies.head(20)
plt.figure(figsize=(8, 8))
sns.scatterplot(x='High', y='Low', data=top_20, hue='Name', palette='Set2')
plt.title('Scatter Plot between High and Low Stock Prices for Top 20 Companies')
plt.xlabel('High Stock Price')
plt.ylabel('Low Stock Price')
plt.show()
```


# Visualization 4: Distribution of Percentage Changes for the top 50 companies


```python
plt.figure(figsize=(10, 6))
sns.histplot(top_50_companies['Chg...'], bins=20, kde=True, color='purple')
plt.title('Distribution of Percentage Changes for Top 50 Companies')
plt.xlabel('Percentage Change')
plt.ylabel('Frequency')
plt.show()
```


# Visualization 5: Pair plot for selected numerical columns for the top 50 companies


```python
selected_columns = ['Last', 'High', 'Low', 'Chg.', 'Chg...']
sns.pairplot(top_50_companies[selected_columns])
plt.suptitle('Pair Plot for Selected Numerical Columns for Top 50 Companies', y=1.02)
plt.show()
```


# Visualization 6: Bar plot for average Last recorded stock prices by industry for the top 10 companies


```python
# Select top 10 companies from the top 50
top_10 = top_50_companies.head(10)

# Extract industry information
top_10['Industry'] = top_10['Name'].apply(lambda x: x.split()[0])

# Visualization: Bar plot for average last recorded stock prices by industry for the top 30 companies
plt.figure(figsize=(12, 6))
sns.barplot(x='Industry', y='Last', data=top_10, palette='muted')
plt.title('Average Last Recorded Stock Prices by Industry for Top 20 Companies')
plt.xlabel('Industry')
plt.ylabel('Average Last Recorded Stock Price')
plt.show()
```

# Visualization 7: Heatmap of correlation matrix for the top 50 companies (using Corr50)

```python
Corr50 = top_50_companies[selected_columns].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(Corr50, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Top 50 Companies')
plt.show()
```
